```python
import pandas as pd
import zipfile

zip_file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_900.zip"
csv_file_name = "API_SP.POP.TOTL_DS2_en_csv_v2_900.csv"

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    try:
        df = pd.read_csv(zip_ref.open(csv_file_name), sep=';', header=0)  # Try semicolon
    except pd.errors.ParserError:
        df = pd.read_csv(zip_ref.open(csv_file_name), sep='\t', header=0)  # Try tab
    except pd.errors.ParserError:

        print("Error: Could not determine the correct delimiter. Check the CSV file format.")

print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate value counts and select the top 5 categories
data_source_counts = df['Data Source,"World Development Indicators",'].value_counts().head(5)

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(data_source_counts.index, data_source_counts.values)
plt.xlabel('Data Source, "World Development Indicators"')
plt.ylabel('Count')
plt.title('Distribution of Top 5 Data Sources')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```
